# Question 1

Which page did user id "" (empty string) NOT visit?

```python
user_log.printSchema()
a = user_log.select(['page', 'userId']).where(user_log.userId == "" ).groupby(user_log.page).count()
# TODO: write your code to answer question 1
a.show()
```

# Question 3

# Question 4

How many songs were played from the most played artist?

```python
a = user_log.select(['artist', 'sessionId']).filter(user_log.artist != "").groupby(user_log.artist).count()
# TODO: write your code to answer question 1
a = a.toPandas()
a.sort_values(by='count', ascending=False)
```
